**03: Rows/index operations on a DataFrame.**

```python
import pandas as pd

# Intiialise data as a dictionary
data = {'Name':['Isaac Newton', 'Albert Einstein', 'Marie Curie', 'Stephen Hawking','Meghnad Saha'],
        'Work':['Gravitation', 'Relativity', 'Radioactivity', 'Black Holes', 'Plasma'],
        'Birth Year':[1643, 1879, 1867, 1942, 1893],
        'Age': [84, 76, 76, 66, 62], 
        'Country':['England','Germany','Poland','England','India']
        }
 
df = pd.DataFrame(data, index=[1,2,3,4,5])
display(df)
df = pd.DataFrame(data, index=range(5,0,-1))
df
```

0. Get row index and labels

```python
# Details of the index 
idx = df.index

# Find the label of index at a given position
posn = 2
label = df.index[posn] 

# Create a list of the index
lst = df.index.tolist()

# Show idx, label, and lst
print(idx,label, lst, sep='\n')
#display(idx,label, lst)
```

1. Rename/change index

```python
print("Method 0: Using `rename` and modify individually")
df1 = df.copy()
df1.rename(index={5:'Five', 1:'One'}, inplace=True)
display(df1)

print("Method 1: Using a list and `index` attribute")
df1 = df.copy()
new_index = ['V','W','X','Y','Z']
df1.index = new_index
display(df1) 

print("Method 2: Using `set_index` and Series/List")
''' Syntax:
    DataFrame.set_index(keys, *, drop=True, append=False, inplace=False, \
       verify_integrity=False)'''
print('  Using a Series ->')
s = pd.Series(['A','B','C','D','E'])
l = ['a','e','i','o','u']
#df1 = df1.set_index(new_index) # doesn't work, needs a column
df1 = df1.set_index([s]) # can use inplace=True option to skip reassignment
display(df1)
print('  Using a list ->')
df1 = df1.set_index([l])
display(df1)


print("Method 3: Using `set_index` and `pandas.Index`:")
indx = pd.Index([0.0, 0.1, 0.2, 0.3, 0.4])
df1.set_index(indx, inplace=True)
display(df1)



```

2. Rearrange index.

```python
df1 = df.copy()
display(df1)

# Reset index
print('Index reset to 0:')
df1.reset_index(drop=True, inplace=True)
display(df1)

#  Shift index 
## Method 0
print('Index will start from 1:')
df1.index = df1.index + 1
display(df1)

# Reset index
print('Index reset to 0:')
df2 = df1.reset_index(drop=True)
display(df1)


# Shift index
## Method 1
print('Index again will start from 1:')
df1.index  = range(1, len(df)+1)
display(df1)


# Set starting index to be 555
## Method 0
print('len(df)=',len(df))
df1 = df2.copy()
df1.index = df1.index + 555
display(df1)

## Method 1
print('Index again will start from 555:')
df1 = df2.copy()
df1.index  = range(555, len(df)+555)
display(df1)

```

3. Order index.

```python
df1 = df.copy()
print('Original DataFrame:')
display(df1)

print('Sorted or ordered DataFrame:')
df1.sort_index(inplace=True)
display(df1)

print('Restored DataFrame by reverse-sorting (descending):')
df1.sort_index(ascending=False, inplace=True)
df1
```

4. Create index from a column.

```python
df2 = df1.copy()

print("Converting the column ['Name'] to index:")
df2.set_index('Name', inplace=True) # default drop = True (=> removes the column)
display(df2) 

print("Changing the index name:")
df2.index.name = 'Scientist'
display(df2)

print("Removing the index name:")
df2.index.name = None
display(df2)

print("Aligning index and column headers")
df2.columns.name = 'Scientist'
df2
```

5. Create a column from index.

```python
df3 = df2.copy()
print("Original DataFrame:")
display(df3)

print("\nCase 0: Create a column from index and then reset index ->")
df3['Scientist'] = df3.index 
df3.reset_index(drop=True, inplace=True)
display(df3)

print("\nCase 1: Reset index and change column name->")
df3 = df2.copy()
df3.reset_index(level=0, inplace=True)
df3 = df3.rename(columns={'index':'Scientist'}) 
df3
```

6. Add a row. 

```python
print('Method 0: Using index name (works for any indexing)')
df4 = df2.copy() # Note df2 doesn't have numbered index
print('Before adding row:')
list_row = ['Eclipses',None, None, 'India']
df4.loc['Brahmagupta'] = list_row
print('After adding row:')
display(df4)  


print('Method 1: Using index position (works for numbered indexing)')
df5 = df3.copy()
print('Before adding row:')
display(df5)
list_row = ['Brahmagupta','Eclipses',None, None, 'India']
df5.loc[len(df3)] = list_row # won't work if index is not from 0 to len(df3)
print('After adding row:')
df5  
#df3.isnull()
```

7. Drop rows.

```python
print("Dropping multiple rows:")
df6 = df4.copy()
df6.drop(['Brahmagupta','Stephen Hawking'], inplace=True)
display(df6)

print("Dropping the last row:")
df6 = df5.copy()
display('df=',df6)
df6.drop(len(df6)-1, inplace=True) 
display(df6)

print("Dropping a range of rows:")
print('Method 0::')
df6 = df5.copy()
df6.drop(range(1,4), inplace=True) 
display(df6)
print('Method 1::')
df6 = df5.copy()
df6.drop(df6.index[3:5], inplace=True)
display(df6)

print("Dropping the rows that have None values:")
df6 = df5.copy()
df6.dropna(inplace=True) 
#df6 = df6[df6.Discount.notnull]
df6



```
