**03: Column operations on DataFrames**

```python
import pandas as pd
```

```python
# intialise data of lists.
data = {'Name':['Isaac Newton', 'Albert Einstein', 'Marie Curie', 'Stephen Hawking','Meghnad Saha'],
        'Work':['Gravitation', 'Relativity', 'Radioactivity', 'Black Holes', 'Plasma'],
        'Birth Year':[1643, 1879, 1867, 1942, 1893],
        'Age': [84, 76, 76, 66, 62], 
        'Country':['England','Germany','Poland','England','India']
        }
 
df = pd.DataFrame(data)
df
```

```python
df1 = df.rename(columns={'Name':'Scientist'})
df1
```

1. Apply a function to a column.

```python
print('Method 0:')
df2 = df1.copy() 
df2['Country'] = df2['Country'].str.lower()
print(df2,'\n') 
df2

print('Method 1:')
df2 = df1.copy()
df2.Country = df1.Country.apply(lambda x: x.upper())
print(df2,'\n')

print('Method 2:')
df2 = df1.copy() 
df2['Country'] = df2['Country'].apply(lambda x: x.upper())
print(df2,'\n')

# Checking df1
print('df1=\n',df1,'\n')

print('Method 3:')
df2 = df1.copy() 
df2['Country'] = df2['Country'].map(lambda x: x.upper())
df2 

```

4. Drop columns.

```python
print("Dropping by mentioning all the columns to be kept (brute force):")
df3 = df2[['Country','Scientist', 'Work', 'Birth Year']]
display(df3)

print("Dropping specific column(s)")
df3 = df2.drop(columns='Age')
print(df3)
df3 = df2.drop(columns=['Age','Work'])
display(df3)
df3 = df2.drop('Birth Year', axis=1) # axis=1 => column
display(df3)

print("Dropping selected multiple columns:")
df3 = df2.drop(['Birth Year','Work'], axis=1) # axis=1 => column
display(df3)

display('df2=',df2)
print("Dropping a range of columns:")
#df3 = df2.drop(df2.columns[range(1,4)])
#df3 = df2.drop(range(1,2), axis = 1)
df3 = df2.drop(df2.iloc[:, 1:4],axis = 1)
df3 = df2.drop(df.loc[:, 'Work':'Age'].columns,axis = 1) # alternative to above
display(df3)
```

6. Insert a column at a specific position.

```python
df4 = df3.copy()
posn = 1
colname = 'Gender'
coldata = ['M','M','F'] 
col2 = ['M']*2
coldata.extend(col2)
print(coldata)
df4.insert(posn, colname, coldata)
df4
```

7. Replace a column.

```python
dob = ['04/01/1643', '14/03/1879', '07/11/1867', '08/01/1942', '06/10/1893']
df5 = df4.copy()  
print('Original df5=\n',df5) 


print("Method 0: By dropping and adding \n (doesn't garantee to replace in the same place):")
df5 = df5.drop(columns = ['Birth Year'])
print(df5)
df5['DOB'] = dob
display(df5) 

print("Method 1: By renaming and updating:")
df5 = df4.rename(columns={'Birth Year':'DOB'})
df5.update({'DOB':dob}) 
display(df5)


print("Method 2: By renaming and assigning:")
print('df4=\n',df4)
df5 = df4.rename(columns={'Birth Year':'DOB'}).assign(DOB=dob) 
#df5 = df5.assign(DOB=dob) 
df5

```

8. Change column positions.

```python
print("Case 0: By brute force ->")
df6 =  df5.copy()
print("Creating a list having the column names in desired order ...") 
lst = ['Country', 'Scientist', 'Gender', 'Work', 'Nobel', 'DOB', 'Age'] 
print('Reindexing the columns ...')
df6 = df6.reindex(columns=lst)
display(df6) 

print("Case 1: By sorting ->")
df6 =  df5.copy()
print("Let us arrange albhabetically.")
df6 = df6.reindex(sorted(df6.columns, reverse=False), axis=1)
display(df6) 
print("Let us arrange albhabetically in the reverse order (alt. method).")
df6.sort_index(axis=1, inplace=True, ascending=False) 
display(df6) 

print("Case 2: By finding the column list ->")
listcol = df6.columns.tolist()
print('List of columns:',listcol)
print("Let us move the last column to the first.")
new_cols=listcol[-1:] + listcol[:-1] 
df6 = df6[new_cols]
df6


```
